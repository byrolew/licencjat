```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from preprocess import preprocess
```

```python
filename = "htt_features_test.pkl"
```

```python
data_dict = preprocess(file_name=filename)
```

```python
X = data_dict["not_shuffled_features"]
y = np.squeeze(data_dict["not_shuffled_labels"])
```

```python
for i, col in enumerate(data_dict["feature_names"]):
    print(i, col)
    plt.figure(figsize=(6, 4))
    bins = np.histogram(X[:, i], bins=43)[1]
    plt.hist(X[y == 1, i], density=True, bins=bins, alpha=0.2, color="C0")
    plt.hist(X[y == 1, i], density=True, bins=bins, alpha=0.9, histtype="step", 
             linewidth=2, color="C0", label="Sygnał")
    plt.hist(X[y == 0, i], density=True, bins=bins, alpha=0.9, histtype="step", hatch="///", edgecolor="C1",
             linewidth=2, color="C1", label="Tło")
    plt.ylabel("Gęstość")
    plt.legend()
    
#     plt.savefig("praca/difference_" + col[6:] + ".png", dpi=500)
    plt.show()
```
