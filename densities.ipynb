```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from preprocess import preprocess
```

```python
filename = "htt_features_test.pkl"
```

```python
data_dict = preprocess(file_name=filename)
```

```python
X = data_dict["not_shuffled_features"]
y = np.squeeze(data_dict["not_shuffled_labels"])
```

```python
for i, col in enumerate(data_dict["feature_names"]):
    print(i, col)
    plt.figure(figsize=(6, 4))
    sns.kdeplot(X[y == 1, i], label="Sygna", shade=True)
    sns.kdeplot(X[y == 0, i], linestyle="dashed", label="To", shade=True)
    plt.xlabel("Warto zmiennej")
    plt.ylabel("Gsto")
    plt.savefig("praca/difference_" + col[6:] + ".png", dpi=500)
    plt.show()
#     break
```
