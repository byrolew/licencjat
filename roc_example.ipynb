```python
import numpy as np
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
y = (np.random.uniform(0, 1, 1000) > 0.5).astype(np.int)
pred_rand = np.random.uniform(0, 1, 1000)
pred_good = y + np.random.normal(0, 0.7, 1000)
x = np.random.uniform(0, 1, 1000)
```

```python
roc_rand = roc_curve(y, pred_rand)
auc_rand = round(roc_auc_score(y, pred_rand), 2)
roc_good = roc_curve(y, pred_good)
auc_good = round(roc_auc_score(y, pred_good), 2)
plt.figure(figsize=(6, 6))
plt.plot(roc_rand[1], roc_rand[0], label="Klasyfikacja losowa [" + str(auc_rand) + "]", linestyle="--")
plt.plot(roc_good[1], roc_good[0], label="Klasyfikacja nielosowa [" + str(auc_good) + "]")
plt.plot(x, x, alpha=0.5, color="grey")
plt.legend()
plt.xlabel("True Positive Rate")
plt.ylabel("False Positive Rate")
plt.savefig("praca/roc_example.png", dpi=500)
plt.show()
```
