```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from preprocess import preprocess
```

```python
filename = "htt_features_test.pkl"
```

```python
data_dict = preprocess(file_name=filename)
```

```python
X = data_dict["not_shuffled_features"]
y = np.squeeze(data_dict["not_shuffled_labels"])
```

```python
for i, col in enumerate(data_dict["feature_names"]):
    print(i, col)
    sns.kdeplot(X[y == 1, i], label="Sygnał", shade=True)
    sns.kdeplot(X[y == 0, i], linestyle="dashed", label="Tło", shade=True)
    plt.xlabel(col[6:])
    plt.ylabel("Gęstość")
    plt.show()
    break
```

```python
i = 20
col = "leg_2_ptWeightedDrIsolation"
sns.kdeplot(X[y == 1, i], label="Sygnał", shade=True)
sns.kdeplot(X[y == 0, i], label="Tło", shade=True, linestyle="dashed")
plt.xlabel(col[6:])
plt.ylabel("Gęstość")
plt.savefig("praca/difference.png", dpi=500)
plt.show()
```
